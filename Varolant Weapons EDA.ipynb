### Modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

sns.set_style('darkgrid')

%matplotlib inline
```

### Initializing the dataframe

```python
df = pd.read_csv('valorant-stats.csv')
df.head(25)
```

### Renaming columns for better analysis

```python
df.rename(columns={'HDMG_0': 'HD: Close', 'BDMG_0': 'BD: Close', 'LDMG_0':'LD: Close', 
                  'HDMG_1':'HD: Medium', 'BDMG_1':'BD: Medium', 'LDMG_1':'LD: Medium', 
                  'HDMG_2':'HD: Long', 'BDMG_2':'BD: Long', 'LDMG_2':'LD: Long'}, 
          inplace=True)

df.head()
```

### Looking for any Null values

```python
df.isnull().sum()
```

### Looking at the Five Points Summary

```python
df.describe()
```

### Cleaning the data

```python
collective_df = df.copy()

collective_df['Avg Head Dmg'] = (df['HD: Close'] + df['HD: Medium'] + df['HD: Long'] / 3).round(2)
collective_df['Avg Body Dmg'] = (df['BD: Close'] + df['BD: Medium'] + df['BD: Long'] / 3).round(2)
collective_df['Avg Leg Dmg'] = (df['LD: Close'] + df['LD: Medium'] + df['LD: Long'] / 3).round(2)

collective_df.drop(columns=['HD: Close', 'HD: Medium', 'HD: Long', 'BD: Close', 'BD: Medium', 'BD: Long', 'LD: Close', 'LD: Medium', 'LD: Long'], 
                   inplace=True)

collective_df.head()
```

### Correlation between variables

```python
plt.figure(figsize=(8, 8))

sns.heatmap(collective_df.corr(), annot=True, cmap='Greens')

plt.show()
```

- Clearly, Head Damage, Body Damage & Leg Damage are almost perfectly correlated.  
    This concludes that, weapons that deal high Head Damage, deal high Body and Leg Damages as well.  
- Weapons with less Mag Capacity have less Fire Rate.  
    Similarly, weapons with less Mag Capacity & less Fire Rate but high Price deal high Damages.

### Relationship between Variables

```python
plt.figure(figsize=(20, 20))

df_grid = sns.PairGrid(data=collective_df)

df_grid.map_diag(sns.histplot, palette='Set1')
df_grid.map_upper(sns.regplot)
df_grid.map_lower(sns.kdeplot)

df_grid.fig.subplots_adjust(top=0.9)
df_grid.fig.suptitle('Relationship between Data Variables', fontweight='bold', fontsize=20)

plt.show()
```

- Large number of Weapons have less Price tag.
- Most of the weapons have high Mag Capacity.
- Most of the weapons have below positional average (median) Damage.

```python
sns.boxplot(data=avgDmg_df, x='Wall Penetration', y='Avg Dmg', palette='Set1')

plt.title('Average Damage over Wall Penetration', fontweight='bold')
plt.xlabel(xlabel='Wall Penetration', fontweight='bold')
plt.ylabel(ylabel='Average Damage', fontweight='bold')
plt.show()
```

- Large number of weapons have high wall penetration.
